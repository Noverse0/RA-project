```python
import pandas as pd
import numpy as np
import datetime
```

```python
data = pd.read_excel('C:/Users/izzac/Desktop/한국부동산원 데이터 병합 완료last.xlsx')
```

```python
# ad 총공급 만들기
data['ad_총공급'] = data['일반공급'] + data['특별공급']

# ad_성명생년중복 만들기
same = data['생년'].astype('string') + data['성명'] # 성명과 생년을 이어 붙여서 사용
same_count = same.value_counts() #성명과 생년 이어 붙인 것 중에 겹치는 것을 찾기
data['ad_생명생년중복'] = 0
for i in range(len(data)):
    data.loc[i,'ad_생명생년중복'] = same_count[same[i]]
    
# ad_성명생년전화중복
same = data['생년'].astype('string') + data['성명'] + data['폰번호'].astype('string') # 성명과 생년, 폰번호를 이어 붙여서 사용
same_count = same.value_counts() #성명과 생년, 폰번호 이어 붙인 것 중에 겹치는 것을 찾기
data['ad_생명생년전화중복'] = 0
for i in range(len(data)):
    data.loc[i,'ad_생명생년중복'] = same_count[same[i]]
    
# ad_변경일자
data['ad_변경일자'] = pd.to_datetime(data['변경일자'].astype('string').str.split(' ').str[0])
# 1900-12-31 이전 날짜 1900-01-00으로 변경
data['ad_변경일자'][data['ad_변경일자'] < datetime.datetime(1900,12,31)] = '1900-01-01'
# na 값처리
data['ad_변경일자'].fillna('1900-01-01')

# ad_행정변경시점
data['ad_행정변경시점'] = pd.to_datetime(data['접수일자']) - pd.to_datetime(data['ad_변경일자'].astype('string').str.split(' ').str[0])

# ad_접수시간
hour = data['접수시간'].astype('string').str.split(':').str[0].astype('int')
minu = data['접수시간'].astype('string').str.split(':').str[1].astype('int')
data['ad_접수시간'] = hour*60 + minu

# ad_신청당첨거주일치여부
data['ad_신청당첨거주일치여부'] = data['신청거주구분'] == data['당첨거주구분']
data['ad_신청당첨거주일치여부'].loc[(data['ad_신청당첨거주일치여부'] == False)] = 0
data['ad_신청당첨거주일치여부'].loc[(data['ad_신청당첨거주일치여부'] == True)] = 1

# ad_부양가족수
data['ad_부양가족수'] = data['부양가족수']
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == '무')] = 0
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == 0)] = 0
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == 1)] = 0
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == 2)] = 0
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == 3)] = 0
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == 4)] = 0
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == 5)] = 0
data['ad_부양가족수'].loc[(data['ad_부양가족수'] == 6)] = 0

# ad_저축가입기간
data['ad_저축가입기간'] = data['저축가입기간']
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '무')] = 0
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '6월 이상 ~ 1년 미만')] = 1
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '1년 이상 ~ 2년 미만')] = 2
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '2년 이상 ~ 3년 미만')] = 3
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '3년 이상 ~ 4년 미만')] = 4
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '4년 이상 ~ 5년 미만')] = 5
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '5년 이상 ~ 6년 미만')] = 6
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '6년 이상 ~ 7년 미만')] = 7
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '7년 이상 ~ 8년 미만')] = 8
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '8년 이상 ~ 9년 미만')] = 9
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '9년 이상 ~ 10년 미만')] = 10
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '10년 이상 ~ 11년 미만')] = 11
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '11년 이상 ~ 12년 미만')] = 12
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '12년 이상 ~ 13년 미만')] = 13
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '13년 이상 ~ 14년 미만')] = 14
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '14년 이상 ~ 15년 미만')] = 15
data['ad_저축가입기간'].loc[(data['ad_저축가입기간'] == '15년 이상')] = 16

# ad_무주택기간
data['ad_무주택기간'] = data['무주택기간']
data['ad_무주택기간'].fillna(0)
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '유주택, 30세미만 미혼인 무주택')] = 1
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '만30세미만, 미혼인 무주택자')] = 1
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '1년 미만')] = 2
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '1년 이상 ~ 2년 미만')] = 3
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '2년 이상 ~ 3년 미만')] = 4
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '3년 이상 ~ 4년 미만')] = 5
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '4년 이상 ~ 5년 미만')] = 6
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '5년 이상 ~ 6년 미만')] = 7
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '6년 이상 ~ 7년 미만')] = 8
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '7년 이상 ~ 8년 미만')] = 9
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '8년 이상 ~ 9년 미만')] = 10
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '9년 이상 ~ 10년 미만')] = 11
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '10년 이상 ~ 11년 미만')] = 12
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '11년 이상 ~ 12년 미만')] = 13
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '12년 이상 ~ 13년 미만')] = 14
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '13년 이상 ~ 14년 미만')] = 15
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '14년 이상 ~ 15년 미만')] = 16
data['ad_무주택기간'].loc[(data['ad_무주택기간'] == '15년 이상')] = 17
```
